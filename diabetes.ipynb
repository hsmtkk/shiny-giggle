```python
import sklearn.datasets
```

```python
dataset = sklearn.datasets.load_diabetes()
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(dataset.data)
```

```python
df.columns = dataset.feature_names
```

```python
df
```

```python
target_df = pd.DataFrame(dataset.target)
```

```python
target_df
```

```python
target_df.columns = ["target"]
```

```python
target_df
```

```python
target_df
```

```python
joined_df = df.join(target_df)
```

```python
joined_df
```

```python
train_df = joined_df[:400]
```

```python
import pycaret.regression as regression
```

```python
regression.setup(data=train_df, target="target")
```

```python
best_model = regression.compare_models()
```

```python
tuned_model = regression.tune_model(best_model)
```

```python
regression.plot_model(tuned_model)
```

```python
regression.plot_model(tuned_model, plot="feature")
```

```python
regression.plot_model(tuned_model, plot="error")
```

```python
final_model = regression.finalize_model(tuned_model)
```

```python
test_df = joined_df[400:]
```

```python
test_df
```

```python
test_predicted = regression.predict_model(final_model, data=test_df.drop(columns=["target"]))
```

```python
test_predicted
```

```python
compare_df = test_df[["target"]].join(test_predicted[["prediction_label"]])
```

compare_df

```python
compare_df
```

```python
compare_df.plot.scatter(x="target", y="prediction_label")
```
