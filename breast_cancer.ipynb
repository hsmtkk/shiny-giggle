```python
import pandas as pd
```

```python
data_df = pd.DataFrame(dataset.data)
```

```python
data_df.columns = dataset.feature_names
```

```python
data_df
```

```python
target_df = pd.DataFrame(dataset.target)
```

```python
target_df.columns = ["target"]
```

```python
target_df
```

```python
full_df = data_df.join(target_df)
```

```python
full_df
```

```python
train_df = full_df[:560]
test_df = full_df[560:]
```

```python
train_df
```

```python
test_df
```

```python
import pycaret.classification as classification
```

```python
classification.setup(data=train_df, target="target")
```

```python
classification.compare_models()
```

```python
rf_model = classification.create_model("rf")
```

```python
tuned_model = classification.tune_model(rf_model)
```

```python
classification.plot_model(rf_model, plot="feature")
```

```python
classification.plot_model(rf_model, plot="confusion_matrix")
```

```python
final_model = classification.finalize_model(rf_model)
```

```python
predicted = classification.predict_model(final_model, test_df.drop(columns=["target"]))
```
