```python
shuffled_df = full_df.sample(frac=1)
```

```python
shuffled_df
```

```python
train_df = shuffled_df[:160]
test_df = shuffled_df[160:]
```

```python
import pycaret.classification as classification
```

```python
classification.setup(data=train_df, target="target")
```

```python
classification.compare_models()
```

```python
ridge_model = classification.create_model("ridge")
```

```python
ridge_model
```

```python
tuned_model = classification.tune_model(ridge_model)
```

```python
classification.plot_model(ridge_model, plot="feature")
```

```python
classification.plot_model(ridge_model, plot="confusion_matrix")
```

```python
final_model = classification.finalize_model(ridge_model)
```

```python
test_df
```

```python
predicted_df = classification.predict_model(final_model, test_df.drop(columns=["target"]))
```
